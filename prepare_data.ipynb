```python
import os
from PIL import Image

def rename_images(folder_path, target_ext=".jpg"):
    exts = [".jpg", ".jpeg", ".png", ".bmp", ".webp"]
    files = [f for f in os.listdir(folder_path)
             if os.path.splitext(f)[1].lower() in exts]

    # Sort để đặt số thứ tự ổn định
    files.sort()

    for idx, filename in enumerate(files, start=1):
        old_path = os.path.join(folder_path, filename)

        # Load ảnh
        img = Image.open(old_path).convert("RGB")

        # New name
        new_name = f"puma_fake_{idx}{target_ext}"
        new_path = os.path.join(folder_path, new_name)

        # Save theo định dạng mới
        img.save(new_path)


        if new_path != old_path:
            os.remove(old_path)

    print("Done")

rename_images(r"D:/khanhsh/data/gallery/puma/fake", target_ext=".jpg")
```
